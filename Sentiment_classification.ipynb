```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from simpletransformers.classification import ClassificationModel, ClassificationArgs
```

```python
path="/content/drive/MyDrive/archive/Test.csv"
```

```python
data=pd.read_csv(path)
data
```

```python
x=data['text']
y=data["label"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
train_df=pd.DataFrame({"review":x_train,"sentimental":y_train})
eval_df=pd.DataFrame({"review":x_test,"sentimental":y_test})
```

```python
model_args=ClassificationArgs()
model_args.num_train_epochs=2
```

```python

model = ClassificationModel(
    "bert", "bert-base-cased", args=model_args
)
```

```python
model.train_model(train_df,acc=accuracy_score)
```

```python
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```
